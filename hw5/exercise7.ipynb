1. Objective : This code is for stock price and volume

2. Input

  * Data file (stockprice_3.xlsx)

3. Output

  * 2019 price graph, price grap with 3MA, price grap with 5MA
  * 2020 price graph, price grap with 3MA, price grap with 5MA
  * 2019 volume graph, volume grap with 3MA, volume grap with 5MA
  * 2020 volume graph, volume grap with 3MA, volume grap with 5MA

```python
#data handling
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_excel('stockprice_3.xlsx')
#data extract
c19 = df['Date'] < 20200101
df19 = df[c19]
c20_1 = df['Date'] < 20210101
c20_2 = df['Date'] > 20191231
df20 = df[c20_1 & c20_2]
```

```python
# Shows in one graph.
data_do_time_19=df19.groupby('Date').mean() 

# for 2019 price 
plt.figure(figsize=(50,15))
plt.subplot(3, 6, 1)
plt.title('2019 Price')
plt.plot(data_do_time_19['Price'], label = 'original')
plt.legend()
# for 2019 price 3MA
nMA = 3
ls=[]
for j in range(nMA,len(data_do_time_19)-nMA): 
    ls.append(data_do_time_19['Price'].iloc[j-nMA:j+nMA].mean())
    
df_mean3=pd.DataFrame({"Date" : data_do_time_19.index.values[nMA:len(data_do_time_19)-nMA], "Mean_Price" : ls })
plt.plot(df_mean3["Date"],df_mean3["Mean_Price"],label = '3MA')
plt.legend()
# for 2019 price 5MA
nMA = 5
ls=[]
for j in range(nMA,len(data_do_time_19)-nMA):
    ls.append(data_do_time_19['Price'].iloc[j-nMA:j+nMA].mean())
    
df_mean5=pd.DataFrame({"Date" : data_do_time_19.index.values[nMA:len(data_do_time_19)-nMA], "Mean_Price" : ls })

plt.plot(df_mean5["Date"],df_mean5["Mean_Price"],label = '5MA')
plt.legend()   
    
# for 2020 price 
data_do_time_20=df20.groupby('Date').mean() 
plt.figure(figsize=(50,15))
plt.subplot(3, 6, 2)
plt.title('2020 Price')
plt.plot(data_do_time_20['Price'],label = 'original')
plt.legend()

# for 2020 price 3MA
nMA = 3
ls=[]
for j in range(nMA,len(data_do_time_20)-nMA): 
    ls.append(data_do_time_20['Price'].iloc[j-nMA:j+nMA].mean())
    
df_mean3=pd.DataFrame({"Date" : data_do_time_19.index.values[nMA:len(data_do_time_20)-nMA], "Mean_Price" : ls })

plt.plot(df_mean3["Date"],df_mean3["Mean_Price"],label = '3MA')
plt.legend()
# for 2020 price 5MA
nMA = 5
ls=[]
for j in range(nMA,len(data_do_time_20)-nMA): 
    ls.append(data_do_time_20['Price'].iloc[j-nMA:j+nMA].mean())
    
df_mean5=pd.DataFrame({"Date" : data_do_time_20.index.values[nMA:len(data_do_time_20)-nMA], "Mean_Price" : ls })

plt.plot(df_mean5["Date"],df_mean5["Mean_Price"],label = '5MA')
plt.legend()

    
plt.show()
```
